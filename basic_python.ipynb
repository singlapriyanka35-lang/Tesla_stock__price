```python
# standard Scaling (z-score Normalization)
# Backup
df_ready = df.copy()
```

```python
# importing important library
from sklearn.preprocessing import StandardScaler

#initialization

Scaler = StandardScaler()

# Making the list of numerical columns

num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']

# Scaling
df_ready[num_cols]  = Scaler.fit_transform(df_ready[num_cols])

df_ready.head()

# mean = 0
# std deviation = 1

# Xscaled = X- mean / std_dev

'''
Standard Scaling makes your data fair for the model.
No feature dominates just because it has large numbers.
'''
```

```python
# Dealing with categorical Columns
# Making a list of categorical columns 
cat_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

#OHE
# Scaling
df_ready  = pd.get_dummies(df_ready,columns = cat_cols, drop_first= True)

# Encoding the target variable

df_ready['deposit'] = df_ready['deposit'].map({'yes':1, 'no' : 0 })

#Replacing False with 0 and True with 1 

df_ready = df_ready.replace({'True':1, 'False' : 0 })

df_ready.head()


```

```python
# seperation the input and output features

X = df_ready.drop('deposit', axis = 1)
y = df_ready['deposit']

# Decision Tree Classifier 
from sklearn.model_selection import train_test_split

# performing Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5 )

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5 )

# Decision Tree Classifier 
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()

dt.fit(X_train , y_train)

#Taking the Predictions
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

```

```python
# Model Evaluattion
from sklearn import metrics
print('Training Accuracy : ', np.round(metrics.accuracy_score(y_train,y_train_pred),3 ))
print('Testing Accuracy : ', np.round(metrics.accuracy_score(y_test,y_test_pred),3 ))

print('Training Precision : ', np.round(metrics.precision_score(y_train,y_train_pred),3 ))
print('Testing Precision : ', np.round(metrics.precision_score(y_test,y_test_pred),3 ))
```

```python
'''
Conclusion ->
As we can see, there is a huge amount of difference b/w performance on training and testing dataset. 

This confirms that the model has overfits the data.

'''
```

```python
# Making the parameter dictionary 
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [1, 2, 3, 4, 5],
    'min_samples_split': [2, 5, 9, 12, 15],
    'min_samples_leaf': [1, 3, 5, 9, 12]
}

# Import
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Using Grid Search CV
grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=param_grid,
    cv=5
)

grid_search.fit(X_train, y_train)

# extracting the best parameters.
print(grid_search.best_params_)

# Creating the best model
dt_best = DecisionTreeClassifier(
    criterion=grid_search.best_params_['criterion'],
    max_depth=grid_search.best_params_['max_depth'],
    min_samples_leaf=grid_search.best_params_['min_samples_leaf'],
    min_samples_split=grid_search.best_params_['min_samples_split']
)

dt_best.fit(X_train, y_train)

# Taking the predictions 
y_train_pred = dt_best.predict(X_train)
y_test_pred = dt_best.predict(X_test)
```

```python
# Model Evaluattion
from sklearn import metrics
print('Training Accuracy : ', np.round(metrics.accuracy_score(y_train,y_train_pred),3 ))
print('Testing Accuracy : ', np.round(metrics.accuracy_score(y_test,y_test_pred),3 ))

print('Training Precision : ', np.round(metrics.precision_score(y_train,y_train_pred),3 ))
print('Testing Precision : ', np.round(metrics.precision_score(y_test,y_test_pred),3 ))
```

```python
'''
Conclusion ->
As we can see, the training and testing performance is  within 5% hence the model is not pverfitted. 
Training and Testing precision both are above 70%. Which means that the model is performing okay.

We can try to optimize the precision further.

This confirms that the model has overfits the data.

'''
```
